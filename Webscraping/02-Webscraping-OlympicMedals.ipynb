```python
import requests
import re
import pandas as pd
import matplotlib.pyplot as plt
```

```python

def stringclean(mystring):
    #regex = re.compile(".*?\((.*?)\)")
    mystring = re.sub("[\(\[].*?[\)\]]", "", mystring)
    mystring = re.sub("[$,]", "", mystring)
    return (mystring)
```

```python
def rettable(url):
    tbls = pd.read_html(url)
    for t in tbls:
        if ('Bronze' in t):
            return (t)
    return ()     
```

```python
cnt = 0
for i in range(1900, 2019, 4):
    print (i)
    #url = "https://en.wikipedia.org/wiki/"  + str(i) + "_in_film"
    url = "https://en.wikipedia.org/wiki/" + str(i) + "_Summer_Olympics_medal_table"
    print (url)
    try:        
        b = rettable(url)
        b['Year'] = str(i)
        if (cnt == 0):
            mlist = b
        else:
            mlist = mlist.append(b)
        cnt += 1
    except:
        pass 
```
