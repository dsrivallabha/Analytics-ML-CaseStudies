```python
import pandas as pd
import matplotlib.pyplot as plt
# https://grouplens.org/datasets/movielens/
```

```python
df = pd.read_csv('movie_ratings.csv')
df.head()
```

```python
# unique users and ratings 
```

```python
print ('unique users', df['userId'].nunique())
print ('unique movies', df['movieId'].nunique())
print ('total records', len(df))
```

```python
# Average no of ratings per user
print ('average ratings per user', len(df)/df['userId'].nunique())
print ('average ratings per movie', len(df)/df['movieId'].nunique())
```

# Frequent Users 

```python
# Filter for frequent raters
dfug = df.groupby(['userId'])['movieId'].agg(['count', 'nunique']).reset_index().sort_values('count', ascending=False)
dfug.head()
```

```python
# each rating is for a unique item  - one user has only one rating per item 
dfug[dfug['count'] != dfug['nunique']]
# each person has a single rating for a product
```

```python
dfug['count'].describe()
```

```python
# frequent users - with 20 to 50 ratings
dffu = dfug[(dfug['count'] > 20) & (dfug['count'] < 50)]
print (len(dffu))
```

```python
freq_users = list(set(dffu['userId']))
len(freq_users)
```

```python
# Ratings 
# Filter for frequent raters
dfrg = df.groupby(['userId'])['rating'].agg(['count', 'mean']).reset_index().sort_values('count', ascending=False)
dfrg.head()
```

```python
plt.scatter(dfrg['count'], dfrg['mean'])
plt.show()
```

# Frequent Movies

```python
# Filter for frequent raters
dfpg = df.groupby(['movieId'])['userId'].agg(['count', 'nunique']).reset_index()
dfpg.head()
```

```python
# each rating is for a unique item  - one user has only one rating per item 
dfpg[dfpg['count'] != dfpg['nunique']]
# each person has a single rating for a product
```

```python
# frequently rated  product - with atleast 50 ratings
dffp = dfpg[(dfpg['count'] > 50) & (dfpg['count'] < 100)]
print (len(dffp))
```

```python
freq_products = list(set(dffp['movieId']))
len(freq_products)
```

# Filter data to these two sets

```python
df1 = df[df['userId'].isin(freq_users)]
df2 = df1[df1['movieId'].isin(freq_products)]
```

```python
print ('no of entries', len(df2))
print ('no of unique users', df2['userId'].nunique())
print ('no of unique movies', df2['movieId'].nunique())
```

```python
# Split this data into train and test
# Hold out some data as test
train=df2.sample(frac=0.8,random_state=0)
test=df2.drop(train.index)
```

```python
# ensure that all test users and products are in train 
test1 = test[test['userId'].isin(list(train['userId']))]
test2 = test1[test1['movieId'].isin(list(train['movieId']))]
print (len(test), len(test2))
```

```python
train.reset_index(drop=True, inplace=True)
train.head()
```

```python
test2.reset_index(drop=True, inplace=True)
test2.head()
```

```python
train.to_csv('Recommendations-train.csv', index=False)
test2.to_csv('Recommendations-test.csv', index=False)
```
