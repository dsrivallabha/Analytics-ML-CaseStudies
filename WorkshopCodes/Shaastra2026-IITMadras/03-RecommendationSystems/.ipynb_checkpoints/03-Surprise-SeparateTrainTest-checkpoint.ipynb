```python
import pandas as pd
from surprise import Dataset, Reader, SVD, KNNBaseline
from surprise import SlopeOne
from surprise.model_selection import train_test_split
from surprise import accuracy
```

# Load Data

```python
traindf = pd.read_csv('Recommendations-train.csv')
traindf.head()
```

```python
traindf[traindf['userId'] == 5]
```

```python
traindf[traindf['movieId'] == 36]
```

```python
traindf.rating.min(), traindf.rating.max()
```

```python
# Sample code: https://www.kaggle.com/code/noorsaeed/simple-recommender-system-with-surprise
reader = Reader(rating_scale=(traindf.rating.min(), traindf.rating.max()))
data = Dataset.load_from_df(traindf[['movieId','userId','rating']],reader)
trainset = data.build_full_trainset()
```

# Singular Value Decomposition

```python
# Use a simple algorithm (SVD)
#model = SVD(n_factors=50, random_state=42)

model = SVD()

# Train the model
model.fit(trainset)
```

```python
# Read test set and predict
```

```python
testdf = pd.read_csv('Recommendations-test.csv')
testdf.head()
```

```python
# Surprise expects a list of (user, item, rating)
testset = list(
    zip(
        testdf['userId'],
        testdf['movieId'],
        testdf['rating']
    )
)
```

```python
testset[:10]
```

```python
# Make predictions
predictions = model.test(testset)
#predictions = model.test(testset)
#print(predictions)
```

```python
pd.DataFrame(predictions)
```

```python
# Add predictions to test dataframe
testdf['predicted_rating'] = [pred.est for pred in predictions]
```

```python
testdf.head()
```

```python
testdf.predicted_rating.describe()
```

```python
# Use a different model 
```

```python
# Use a simple algorithm (SVD)
model2 = KNNBaseline()

# Train the model
model2.fit(trainset)
```

```python
# Make predictions
predictions2 = model2.test(testset)
#predictions = model.test(testset)
#print(predictions)
```

```python
# Add predictions to test dataframe
testdf['KNN_preds'] = [pred.est for pred in predictions2]
```

```python
testdf.head()
```

# Compare errors

```python
testdf['SVDError'] = abs(testdf['predicted_rating'] - testdf['rating'])
testdf['KNNError'] = abs(testdf['KNN_preds'] - testdf['rating'])
```

```python
import matplotlib.pyplot as plt 
```
