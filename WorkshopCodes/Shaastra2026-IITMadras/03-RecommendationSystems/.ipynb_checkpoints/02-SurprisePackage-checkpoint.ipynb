```python
import pandas as pd
from surprise import Dataset, Reader, SVD
from surprise import SlopeOne
from surprise.model_selection import train_test_split
from surprise import accuracy
```

```python
# pip install scikit-surprise
# pip install "numpy<2.0" 
# Sample code: https://www.kaggle.com/code/noorsaeed/simple-recommender-system-with-surprise
```

# Load Data

```python
df = pd.read_csv('Recommendations-train.csv')
df.head()
```

```python
#df[df['userId'] == 5]
df[df['movieId'] == 36]
```

```python
rating_min, rating_max = df['rating'].min(), df['rating'].max()
rating_min, rating_max
```

```python
# Sample code: https://www.kaggle.com/code/noorsaeed/simple-recommender-system-with-surprise
reader = Reader(rating_scale=(rating_min, rating_max))
data = Dataset.load_from_df(df[['movieId','userId','rating']],reader)
```

```python
data
```

```python
trainset, testset = train_test_split(data, test_size=0.2)
```

```python
# Singular Value Decomposition
```

```python
# Use a simple algorithm (SVD)
model = SVD()

# Train the model
model.fit(trainset)
```

```python
# Make predictions
predictions = model.test(testset)

# Evaluate performance
rmse = accuracy.rmse(predictions)
```

```python
# Predict a specific user-movie rating
# Predict rating for a given movie 
pred = model.predict(uid=1, iid=103)
print(f"Predicted rating: {pred.est:.2f}")
```

```python
rdf1 = pd.DataFrame(predictions)
rdf1
```

```python
rdf1.est.describe()
```

# SlopeOne

```python
# Initialize SlopeOne model
model2 = SlopeOne()

# Train the model
model2.fit(trainset)

# Make predictions
predictions2 = model2.test(testset)
```

```python
accuracy.rmse(predictions2)
```

```python
# Predict a specific user-movie rating
pred = model.predict(uid=1, iid=103)
pred
#print(f"Predicted rating: {pred.est:.2f}")
```

```python
rdf2 = pd.DataFrame(predictions2)
rdf2
```

```python
rdf2.est.describe()
```
