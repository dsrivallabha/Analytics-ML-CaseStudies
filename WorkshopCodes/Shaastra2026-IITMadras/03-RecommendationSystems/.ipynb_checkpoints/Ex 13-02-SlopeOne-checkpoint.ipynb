# Slope One 

```python
from collections import defaultdict

class SlopeOne:
    def __init__(self):
        # deviation[i][j] = average rating difference between item i and j
        self.deviation = defaultdict(lambda: defaultdict(float))
        # freq[i][j] = number of users who rated both i and j
        self.freq = defaultdict(lambda: defaultdict(int))
        self.items = set()
        self.dataframe = None

    def fit(self, df):
        """
        data: data frame
        """
        assert set(df.columns) == set(['user', 'item', 'rating']), 'cols do not match user, item, rating'
        
        # store training dataframe 
        self.dataframe = df
        
        # Convert DataFrame to dict: user -> {item: rating}
        data = (df.groupby('user').apply(lambda x: dict(zip(x['item'], x['rating']))).to_dict())
        
        for user, ratings in data.items():
            for i, r_i in ratings.items():
                self.items.add(i)
                for j, r_j in ratings.items():
                    if i != j:
                        self.freq[i][j] += 1
                        self.deviation[i][j] += (r_i - r_j)

        # compute average deviation
        for i in self.deviation:
            for j in self.deviation[i]:
                self.deviation[i][j] /= self.freq[i][j]

    def predict_rating(self, user_ratings):
        """
        user_ratings: dict of item -> rating for the target user
        Returns: dict of item -> predicted rating
        """
        predictions = defaultdict(float)
        frequencies = defaultdict(int)

        for j in self.items:
            if j not in user_ratings:
                for i, r_i in user_ratings.items():
                    if j in self.deviation[i]:
                        freq = self.freq[i][j]
                        predictions[j] += freq * (r_i - self.deviation[i][j])
                        frequencies[j] += freq

        # finalize averages
        return {j: (predictions[j] / frequencies[j]) if frequencies[j] > 0 else None
                for j in self.items if j not in user_ratings}

    def predict(self, df):
        assert set(df.columns) == set(['user', 'item', 'rating']), 'cols do not match user, item, rating'
        # get predict dictionary
        pred_dict = {}
        for user in list(set(df['user'])):
            dff = self.dataframe[self.dataframe['user'] == user]
            user_ratings = dict(zip(dff.item, dff.rating))
            preds = self.predict_rating(user_ratings)
            pred_dict[user] = preds
    
        predlist = []
        for k in pred_dict.keys():
            rating_dict = pred_dict[k]
            for item, rating in rating_dict.items():
                if rating is not None:
                    predlist.append([k, item, rating])
        preddf = pd.DataFrame(predlist)
        preddf.columns = ['user', 'item', 'predicted_rating']
        # fill unknown ratings with zero
        preddf.fillna(0, inplace=True)

        return (preddf)

        
```

```python
# Example dataframe
data = pd.read_csv('Data/ExampleRecommendationData.csv')
print (data) 
data.columns = ['user', 'item', 'rating']

model1 = SlopeOne()
model1.fit(data)
```

```python
preds = model1.predict(data)
```
