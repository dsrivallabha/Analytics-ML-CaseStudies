## 01-02 Basic cleaning

Review text contains non alphanumeric characters, special characters etc. These cannot be interpreted properly, and hence should be removed for seniment analysis

```python
# Tokenizing and stop word removal
inp_text = """This is a sample sentence, showing off </ br> the stop words filtration -- __            ( ) ."""
word_tokens = word_tokenize(inp_text.lower())
print("word tokens", word_tokens)
```

```python
# Check all special characters and remove them in tokens

# join all reviews into a list
review_list = ["".join(row) for row in df["review"][:5]]

# combine this as a string
combined = " ".join(x for x in review_list)

# get word tokens
combined_tokens = word_tokenize(combined.lower())
print(combined_tokens)
```

```python
s = '!!'
clean = ''.join(c for c in s if s.isalnum())
len(clean)
```

```python
def nacdictionary(token_list):
    """
    dictionary to identify special characters in tokens, to filter them
    """
    nac_dict = {}
    non_alphanumeric_count = 0

    # check if token in entirely non alpha numeric
    for token in token_list:        
        clean = ''.join(c for c in token if c.isalnum())
        if len(clean) == 0:
            if token in nac_dict.keys():
                nac_dict[token] += 1
            else:
                nac_dict[token] = 1

    # Filter + sort (descending)
    filtered_sorted = {
        k: v
        for k, v in sorted(nac_dict.items(), key=lambda x: x[1], reverse=True)
        if v >= 1
    }

    return filtered_sorted


# Example usage
token_list = ["apple", ">", "!!", "!!", "!!", "!!", "!!", """it's"""]
nacdictionary(token_list)
```

# 02 - Examine word frequencies in positive and negative reviews

## 02-01-Word processing

```python
# Download stopwords from NLTK
# nltk.download("punkt_tab")
nltk.download("stopwords")
print(stopwords.words("english"))
stop_words = set(stopwords.words("english"))
```

```python
# Tokenizing and stop word removal
inp_text = """This is a sample sentence, showing off </ br> the stop words filtration -- __            ( ) ."""
word_tokens = word_tokenize(inp_text.lower())
fs = [w for w in word_tokens if not w in stop_words]
print("word tokens", word_tokens)
print("filtered sentence", fs)
```

```python
ps = PorterStemmer()
def stopwordremoval(inp_text):

    # break words into tokens
    word_tokens = word_tokenize(inp_text.lower())

    # remove stop words
    fs1 = [w for w in word_tokens if not w in stop_words]

    # remove custom list
    customlist = [">", "<", "/", ".", ",", "br", ")", "(", "!", "--"]
    fs = [w for w in fs1 if w not in customlist]

    # stem the words
    sw = [ps.stem(w) for w in fs]

    return sw
```

```python
# Tokenizing and stop word removal
inp_text = (
    """This is a sample sentence, showing off </ br> the stop words filtration."""
)
word_tokens = stopwordremoval(inp_text)
print("word tokens", word_tokens)
```

```python
df["filtered_words"] = df["review"].apply(lambda x: stopwordremoval(x))
```

```python
only_neg_words_sorted = sorted(
    only_neg_words.items(), key=lambda item: item[1], reverse=True
)
print(only_neg_words_sorted[:50])
```

# 03 - Examine length of positive and negative reviews

```python
def count_chars(text_string):
    x = len(text_string.replace(" ", ""))
    return x


# Example usage
my_string = "Hello, World! 123 #$%^"
# my_string = 'pine apple'
count = count_chars(my_string)
print(count)
```

```python
def count_non_alphanumeric(text_string):
    """
    Counts the number of non-alphanumeric characters in a given string.

    Args:
        text_string: The input string to analyze.

    Returns:
        The count of non-alphanumeric characters.
    """
    non_alphanumeric_count = 0
    for char in text_string.replace(" ", ""):
        if not char.isalnum():
            non_alphanumeric_count += 1
    return non_alphanumeric_count


# Example usage
my_string = "Hello, World! 123 #$%^"
count = count_non_alphanumeric(my_string)
print(f"The number of non-alphanumeric characters is: {count}")
```

```python
def plot_wordcloud(text):
    # create the word cloud
    wordcloud = WordCloud(width=800, height=400, background_color="white").generate(
        text
    )
    # Plot the word cloud using matplotlib
    plt.figure(figsize=(10, 6))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")  # Hide the axis
    plt.show()
    return ()
```

```python
def plot_wordcloud_dict(dicts):
    # create the word cloud
    wordcloud = WordCloud(width=800, height=400, background_color="white").generate_from_frequencies(dicts)
    #wordcloud = WordCloud().generate_from_frequencies(dicts)
    # Plot the word cloud using matplotlib
    plt.figure(figsize=(10, 6))
    plt.imshow(wordcloud)
    plt.axis("off")  # Hide the axis
    plt.show()
    return ()
```

```python
def req_tokens(wordlist):
    newlist = []
    for w in wordlist:
        if w in only_pos_words.keys():
            newlist.append(w)
        if w in only_neg_words.keys():
            newlist.append(w)
    return (newlist)

def bigrams(wordlist):
    bigram_list = list(ngrams(wordlist, 2))
    return bigram_list
```

```python
df['imp_words'] = df["filtered_words"].apply(lambda x: req_tokens(x))
df.head()
```

```python
df['bgl'] = df['imp_words'].apply(lambda x: bigrams(x))
```
