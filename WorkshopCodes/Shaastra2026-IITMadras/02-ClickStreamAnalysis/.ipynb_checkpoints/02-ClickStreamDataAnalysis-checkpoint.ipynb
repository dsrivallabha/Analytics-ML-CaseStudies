# Session statistics

```python
dfg = df[['session', 'buy']].groupby('session').agg(['count', 'sum']).reset_index()
dfg.columns = ['session', 'buy_count', 'buy_sum']
dfg
```

```python
# Sessions purchase ratio
dfg['purchase_ratio'] = dfg['buy_sum']/dfg['buy_count']
```

```python
plt.hist(dfg['purchase_ratio'])
plt.xlabel('Purchase ratio')
plt.ylabel('Count')
plt.show()
```

```python
plt.scatter(dfg['buy_count'], dfg['purchase_ratio'])
plt.show()
```

# Product Statistics

```python
dfp = df[['product_id', 'buy', 'campaign']].groupby('product_id').agg(['count', 'sum']).reset_index()
dfp.columns = ['product_id', 'view_count', 'buy_sum', 'campaign_count', 'campaign_sum']
dfp.sort_values('view_count',ascending=False, inplace=True)
dfp.head(20)
```

```python
dfp['purchase_view_ratio'] = dfp['buy_sum']/dfp['view_count']
dfp['campaign_view_ratio'] = dfp['campaign_sum']/dfp['view_count']
```

```python
plt.scatter(dfp['campaign_view_ratio'], dfp['purchase_view_ratio'])
plt.xlabel('campaign to view ratio')
plt.ylabel('purchase to view ratio')
plt.show()
```

```python
plt.scatter(dfp['campaign_view_ratio'], dfp['purchase_view_ratio'])
plt.xlabel('campaign to view ratio')
plt.ylabel('purchase to view ratio')
plt.xlim(0.08, 0.18)
plt.ylim(0.15, 0.25)
plt.show()
```

# Generate Channel level insights

```python
dfc = df[['channel', 'buy', 'campaign']].groupby('channel').agg(['count', 'sum']).reset_index()
dfc.columns = ['channel', 'view_count', 'buy_sum', 'campaign_count', 'campaign_sum']
dfc
```

```python
dfc['campaign_channel_ratio'] = dfc['campaign_sum']/dfc['view_count']
dfc
```
