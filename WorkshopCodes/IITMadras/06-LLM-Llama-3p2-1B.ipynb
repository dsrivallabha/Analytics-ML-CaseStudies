```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForCausalLM
```

```python
tokenizer = AutoTokenizer.from_pretrained("meta-llama/Llama-3.2-1B")
model = AutoModelForCausalLM.from_pretrained("meta-llama/Llama-3.2-1B")
```

```python
type(model)
```

```python
def text_completion(inp_text, max_new_tokens=50):
  inputs = tokenizer([inp_text], return_tensors="pt")
  outputs = model.generate(**inputs, max_new_tokens=max_new_tokens, return_dict_in_generate=True, output_scores=True)
  transition_scores = model.compute_transition_scores(outputs.sequences, outputs.scores, normalize_logits=True)
  input_length = 1 if model.config.is_encoder_decoder else inputs.input_ids.shape[1]
  generated_tokens = outputs.sequences[:, input_length:]
  strr = ' '
  for gg in generated_tokens:
    strr += tokenizer.decode(gg) + ' '

  return (strr)
```

```python
inp_text = 'Complete the poem correctly. Twinkle twinkle little star '
text_completion(inp_text)
```

```python
inp_text = 'Twinkle twinkle'
text_completion(inp_text)
```

```python
inp_text = 'What is the capital of Japan'
text_completion(inp_text)
```
